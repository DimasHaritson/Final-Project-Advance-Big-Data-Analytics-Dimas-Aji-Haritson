```python
import pandas as pd
```

```python
adidas_sales = pd.read_csv("Adidas Sales.csv")
adidas_sales.head()
```

```python
adidas_sales.info()
```

```python
adidas_sales.describe()
```

```python
# Melihat nilai Null pada data
null_values = adidas_sales.isnull().sum()
null_values
```

Pengecekan nilai null sangat penting dalam analisis data karena berkontribusi pada tercapainya tingkat akurasi yang tinggi. Proses ini membantu memastikan bahwa dataset tidak memiliki data yang hilang atau tidak lengkap, yang bisa menyebabkan ketidakakuratan dalam analisis. Setelah melakukan pengecekan nilai null, ditemukan 2 nilai null pada kolom "price per unit."

```python
duplikat = adidas_sales.duplicated()
duplikat
```

Pengecekan nilai duplikat sangat penting dalam analisis data karena dapat memengaruhi hasil analisis dan mengurangi akurasi data. Nilai duplikat bisa terjadi akibat kesalahan dalam pengumpulan atau pemrosesan data, dan jika tidak ditangani, dapat menyebabkan bias dalam analisis atau perhitungan statistik. Setelah melakukan pengecekan terhadap nilai duplikat, tidak ditemukan data yang duplikat dalam dataset yang saya gunakan. Hal ini memastikan bahwa dataset bersih dari entri berulang yang bisa mengganggu proses analisis. Dengan demikian, analisis dapat dilakukan dengan lebih akurat dan memberikan hasil yang lebih dapat diandalkan.

```python
# Menghapus baris yang memiliki nilai null
adidas_sales = adidas_sales.dropna()
```

```python
# Melihat nilai Null pada data
null_values = adidas_sales.isnull().sum()
null_values
```

```python
# Menyimpan dataframe yang sudah dibersihkan ke dalam file CSV baru
adidas_sales.to_csv('Adidas Sales Cleansed.csv', index=False)
print("Data telah disimpan ke CSV")
```
